```python
from causaldag import dci, dci_stability_selection
from causaldag.datasets import create_synthetic_difference
import numpy as np
import itertools as itr

X1, X2, true_difference = create_synthetic_difference(nnodes=9, nsamples=20000)
p = X1.shape[1]

difference_matrix = dci(X1, X2, difference_ug=list(itr.combinations(range(p), 2)), verbose=True)
ddag_edges = set(zip(*np.where(difference_matrix != 0)))
print(true_difference)
print(ddag_edges)
```
