```python
import numpy as np
from causaldag import dci
import os
import itertools as itr
```

```python
X1 = np.loadtxt('../data/dci_X1.txt')
X2 = np.loadtxt('../data/dci_X2.txt')
```

```python
nodes = list(range(X1.shape[1]))
difference_matrix = dci(X1, X2, difference_ug=set(itr.combinations(nodes, 2)), nodes_cond_set=set(nodes), verbose=True)
ddag_edges = set(zip(*np.where(difference_matrix != 0)))
```
